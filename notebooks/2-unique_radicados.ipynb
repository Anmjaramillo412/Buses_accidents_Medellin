```python
import pandas as pd
import config
```

```python
vehiculos_2013_2023 = pd.read_excel(config.BASE_DATOS_INICIAL)
print("Data shape before filtering: ", vehiculos_2013_2023.shape)
vehiculos_2013_2023.head()
```

```python
solo_buses_2013_2023 = vehiculos_2013_2023[
    vehiculos_2013_2023["CLASE_VEHICULO"].isin(config.TYPE_BUSES)
]
print("Data shape after filtering: ", solo_buses_2013_2023.shape)
solo_buses_2013_2023.head()
```

```python
solo_buses_unique_radicados = pd.DataFrame(
    solo_buses_2013_2023["NRO_RADICADO"].unique()
)
print("Total of accidents involving buses: ",len(solo_buses_unique_radicados))
```

```python
solo_buses_unique_radicados.to_excel(config.LISTA_UNICOS_RADICADOS_BUSES, index=False)
```
